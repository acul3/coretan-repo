```python
def num_there(s):
    return any(i.isdigit() for i in s)
```

```python
def filter_url_by_tld(ls):
    pol = []
    for m in ls:
        url = m.split('.')[-1]
        list_url = ['com','id','org']
        if url in list_url and not num_there(m):
            pol.append(m)
    return pol
```

```python
def check_oscar_by_url(all_name,site_name):
    pol = []
    for o in tqdm(all_name):
        with gzip.open(open(o, "rb"), "rt", encoding="utf-8") as f:
            for line in f:
                if len(pol) > 30:
                    break
                if line:
                    example = json.loads(line)
                    lis = urlparse(example['headers']['warc-target-uri']).netloc
                    if lis == site_name:
                        pol.append({"data":example,"filepath":o})
    #sorted_pol = list(set(pol))
    return pol
```
