```python
import math
def maxPrimeFactor(n):
   # number must be even
   while n % 2 == 0:
      max_Prime = 2
      n /= 1
   # number must be odd
   for i in range(3, int(math.sqrt(n)) + 1, 2):
      while n % i == 0:
         max_Prime = i
         n = n / i
   # prime number greator than two
   if n > 2:
      max_Prime = n
   return int(max_Prime)
```

```python
import gzip
import json
def create_json(lis,name):
    json_lines = [json.dumps(l) for l in lis]
    json_data = '\n'.join(json_lines)
    filename = "conllu_clean/" + str(name) + ".jsonl"
    with gzip.open(filename, 'wt', encoding='UTF-8') as fout:
        fout.write(json_data)
```

```python
create_json(lis,'cc12m_indo1')
```
