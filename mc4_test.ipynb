```python
def check_mc4_by_url(all_name,site_name):
    pol = []
    for o in tqdm(all_name):
        with gzip.open(open(o, "rb"), "rt", encoding="utf-8") as f:
            for i,line in enumerate(f):
                if len(pol) > 40:
                    break
                if line:
                    example = json.loads(line)
                    lis = urlparse(example.get('url')).netloc
                    if lis == site_name:
                        pol.append({'text':example['text'],'url':example.get('url'),'filename':o,'line_number':i})
    #sorted_pol = list(set(pol))
    return pol
```

```python
import gzip
import json
from tqdm.notebook import tqdm,trange
def create_json(lis,name):
    json_lines = [json.dumps(l) for l in lis]
    json_data = '\n'.join(json_lines)
    filename = "oscar_id_clean/" + str(name) + ".jsonl"
    with gzip.open(filename, 'wt', encoding='UTF-8') as fout:
        fout.write(json_data)     
```

```python
def filter_mc4(all_name):
    content = []
    duplicate = []
    for o in tqdm(all_name,desc='Filter Oscar',position = 0, leave = True):
        pol = []
        with gzip.open(open(o, "rb"), "rt", encoding="utf-8") as f:
            for i,line in tqdm(enumerate(f),position = 1, leave=True):
                if line:
                    example = json.loads(line)
                    if example['text'] not in content:
                        content.append(example['text'])
                        #pol.append(example)
                    else :
                        duplicate.append(example['text'])
                        
        #print('saving')
        #create_json(pol,o.split('/')[-1])
    return duplicate,content
```
